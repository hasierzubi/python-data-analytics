```python
import pandas as pd
```

```python
# Ruta del archivo Excel (corrigiendo las barras)
archivo_excel = r"C:\Users\hzubi\OneDrive\Escritorio\airbus\leer_plot\random_updown_data_text.xlsx"

df = pd.read_excel(archivo_excel, dtype=str)

# Verificar los tipos de datos
print(df.info())
```

```python
df.info()
```

```python
# Añadir una nueva columna con valores secuenciales del 1 al 120
df["SequentialNumber"] = range(1, len(df) + 1)

# Verificar los cambios
print(df.head())
```

```python
df.info()
```

```python
# Añadir una nueva columna con valores secuenciales de 2 en 2
df["EvenSequentialNumber"] = range(2, 2 * (len(df) + 1), 2)
```

```python
df.head()
```

```python
df.info()
```

```python
# Convertir columnas a tipo numérico
df["RandomNumber"] = pd.to_numeric(df["RandomNumber"])
df["UpDown"] = pd.to_numeric(df["UpDown"])

# Verificar los cambios
print(df.info())
```

```python
# Eliminar la columna "UpDownPlus100"
df.drop(columns=["UpDownPlus100"], inplace=True)

# Verificar los cambios
print(df.info())
```

```python
import matplotlib.pyplot as plt

# Convertir SequentialNumber a numérico por si acaso
df["SequentialNumber"] = pd.to_numeric(df["SequentialNumber"])

# Crear la figura
plt.figure(figsize=(10, 5))

# Graficar RandomNumber y UpDown
plt.plot(df["SequentialNumber"], df["RandomNumber"], label="RandomNumber", marker="o")
plt.plot(df["SequentialNumber"], df["UpDown"], label="UpDown", marker="s")

# Personalizar el gráfico
plt.xlabel("Sequential Number")
plt.ylabel("Values")
plt.title("Visualization of RandomNumber and UpDown")
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convertir SequentialNumber a numérico por si acaso
df["SequentialNumber"] = pd.to_numeric(df["SequentialNumber"])

# Crear la figura
plt.figure(figsize=(12, 6))

# Graficar todas las columnas numéricas excepto SequentialNumber
for column in df.columns:
    if column != "SequentialNumber":  # Evitar graficar el eje X
        plt.plot(df["SequentialNumber"], df[column], label=column, marker="o")

# Personalizar el gráfico
plt.xlabel("Sequential Number")
plt.ylabel("Values")
plt.title("Visualization of Multiple Columns")
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
